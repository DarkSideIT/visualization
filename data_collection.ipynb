```python
from bs4 import BeautifulSoup
import requests
```

```python
import pandas as pd
```

```python
html = requests.get('https://genshin-info.ru/wiki/personazhi/').text
soup = BeautifulSoup(html, 'html')
```

```python
items = []
index = 1
traveler_elements = ['Анемо', 'Гео', 'Электро', 'Дендро', 'Гидро']
for item in soup.find_all('div', {'class':'itemList__item'}):
  index_character = index
  name = 'Genshin Impact'
  hero = item.find('div', {'class': 'itemcard__name'}).text.strip()
  if hero == '':
    hero = "Путешественник " + traveler_elements.pop()
  url_raw = item.find('img', {'data-src': True})['data-src']
  url = 'https://genshin-info.ru' + url_raw
  group = 'Games'
  items.append({'index_character': index_character, 'name': name, 'hero': hero, 'url': url, 'group': group})
  index += 1
pd.DataFrame(items)
```

```python
df = pd.DataFrame(items)
df.to_csv('genshin_impact.csv', index = False)
```
